```python
def extraer_oscars(url_wikipedia,nombre_archivo):
    res_wikipedia = requests.get(url_wikipedia)
    
    if res_wikipedia.status_code != 200:
        raise Exception (f"El código del error es :{res_wikipedia.status_code}")
    
    sopa_peliculas = BeautifulSoup(res_wikipedia.content, "html.parser")
    tablas = sopa_peliculas.findAll("table")

    mi_tabla = tablas[2]

    lineas = mi_tabla.findAll("tr")

    resultados = []

    for fila in lineas[83:]:
        columnas = fila.findAll("td")
        
        if len(columnas) >= 5:
            
            fecha = columnas[1].text.strip()
            fecha_limpia = re.sub(r"\s+", " ", fecha)  #
            fecha_limpia = re.sub(r"(\w+)(de)", r"\1 de", fecha_limpia) 
            mejor_pelicula = columnas[2].text.strip()
            mejor_director = columnas[3].text.strip()
            mejor_actor = columnas[4].text.strip()
            mejor_actriz = columnas[5].text.strip()

            resultados.append((fecha_limpia, mejor_pelicula, mejor_director, mejor_actor, mejor_actriz))

    columnas_df = ["Fecha", "Mejor Película", "Mejor Director", "Mejor Actor", "Mejor Actriz"]
    df_oscars = pd.DataFrame(resultados, columns=columnas_df)

    df_oscars.to_csv(nombre_archivo, index=False)

    return df_oscars
```

```python
url_wikipedia = "https://es.wikipedia.org/wiki/Premios_%C3%93scar"
```

```python
extraer_oscars(url_wikipedia,"oscars.csv")
```
