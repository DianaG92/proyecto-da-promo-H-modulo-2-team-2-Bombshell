Creamos la función de extraer datos para el enlace url wikipedia y el fichero csv 

```python
def extraer_oscars(url_wikipedia,nombre_archivo): #la primera parte de la función es la variable de que página tenemos que extraer datos, la segunda parte es el nombre del fichero donde vamos a almacenar los datos
    res_wikipedia = requests.get(url_wikipedia) #Solicitud para obtener los datos
    
    if res_wikipedia.status_code != 200:
        raise Exception (f"El código del error es :{res_wikipedia.status_code}") #le ponemos que si la request ha salido mal, nos diga cuál es el error.
     
    sopa_peliculas = BeautifulSoup(res_wikipedia.content, "html.parser") #empezamos con beautifulsoup
    tablas = sopa_peliculas.findAll("table") #buscamos todas las tablas de la pagina

    mi_tabla = tablas[2] #despues de haber verificado fuera de la función cuál era la tabla correcta, creamos una nueva variable que contenga solamente la tabla necesaria.

    lineas = mi_tabla.findAll("tr") #Le decimos que nos traiga los datos de todas las lineas de la tabla

    resultados = [] #creamos una lista vacia para poder añadir los datos extraídos más adelante

    for fila in lineas[83:]: #empezamos a partir de la linea 83 para obtener los datos simplemente a partir del año 2010
        columnas = fila.findAll("td") #sacamos los datos dentro de las columnas 
        
        if len(columnas) >= 5: #cogemos simplemente las primeras 5 columnas ya que no necesitamos los datos de las otras columnas
            
            fecha = columnas[1].text.strip() #entramos en la primera columna y borramos los espacios innecesarios para cada columna
            fecha_limpia = re.sub(r"\s+", " ", fecha)  
            #buscamos cualquier espacio multiple para convertilo a uno solo. El año de estreno estaba en la linea con un /n despues de un espacio.
            #Intentamos usar un slicing pero alargaba mucho el codigo así que lo resolvimos con regex.
            fecha_limpia = re.sub(r"(\w+)(de)", r"\1 de", fecha_limpia) 
            #buscamos dónde aparecen los "de" en la fecha, que puede ser despues de un espacio o despues de unos caracteres, y luego lo reemplazamos para que siempre haya un espacio antes de la palabra "de"
            mejor_pelicula = columnas[2].text.strip()
            mejor_director = columnas[3].text.strip()
            mejor_actor = columnas[4].text.strip()
            mejor_actriz = columnas[5].text.strip()

            resultados.append((fecha_limpia, mejor_pelicula, mejor_director, mejor_actor, mejor_actriz)) #añadimos los datos extraídos a nuestra lista vacía

    columnas_df = ["Fecha", "Mejor Película", "Mejor Director", "Mejor Actor", "Mejor Actriz"] #creamos el dataframe
    df_oscars = pd.DataFrame(resultados, columns=columnas_df)

    df_oscars.to_csv(nombre_archivo, index=False) #convertimos nuestro df a fichero csv sin incluir los índices de las filas

    return df_oscars
```

```python
url_wikipedia = "https://es.wikipedia.org/wiki/Premios_%C3%93scar" #creamos la variable url_wikipedia que llamamos en nuestra función
```

```python
extraer_oscars(url_wikipedia,"Fase_4_datos_oscars.csv") #hacemos uso de la función para extraer la información
```
